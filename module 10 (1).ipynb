#### 13.What is the Central Limit Theorem(CLT), why is it critical in statistics?

**Answer:-** The Central Limit Theorem (CLT) is one of the most powerful ideas in statistics because it explains why the normal distribution shows up everywhere.<br>
If you take random samples of size n from any population (with finite mean and variance):<br>
‚Ä¢The sampling distribution of the sample mean will be approximately normal<br>
‚Ä¢This becomes more accurate as the sample size n increases<br>
‚Ä¢Regardless of the population‚Äôs original distribution (skewed, uniform, bimodal, etc.)<br>
**Why the CLT is critical in statistics:** <br>
**1. Makes normal-based methods possible:** <br>
Many statistical tools assume normality. Thanks to the CLT, we can still use them even when the population itself is not normal, as long as the sample size is large enough.<br>
**2. Enables statistical inference:** <br>
The CLT is the backbone of:<br>
‚Ä¢Confidence intervals<br>
‚Ä¢Hypothesis testing<br>
‚Ä¢p-values<br>
‚Ä¢Without it, much of inferential statistics would fall apart.

#### 17.What are point estimates and interval estimates in statistics?

#### 21.What are the assumptions for applying the Central Limit Theorem?

**Answer:-** The Central Limit Theorem (CLT) says that, under certain conditions, the sampling distribution of the sample mean becomes approximately normal as the sample size gets large. To apply it properly, these assumptions/conditions are typically required:<br>
**1. Independence** <br>
‚Ä¢The observations must be independent of each other.<br>
‚Ä¢In practice, this usually comes from random sampling or random assignment.<br>
‚Ä¢If sampling without replacement, a common rule of thumb is the 10% condition: the sample size should be no more than 10% of the population.<br>
**2. Sample Size is Large Enough** <br>
There‚Äôs no single ‚Äúmagic‚Äù number, but:<br>
‚Ä¢If the population distribution is approximately normal, even a small sample size works.<br>
‚Ä¢If the population is skewed or non-normal, a larger sample is needed (often n ‚â• 30 is used as a rough guideline).<br>
‚Ä¢Heavily skewed distributions or those with outliers may require much larger samples.<br>
**3. Finite Mean and Variance** <br>
‚Ä¢The population must have a finite mean and finite variance.<br>
Extremely heavy-tailed distributions (like some pathological cases) can violate this.
