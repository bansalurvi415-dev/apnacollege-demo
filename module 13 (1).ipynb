#### 1.What is hypothesis testing in statistics?

**Answer:-** Hypothesis testing is a structured way to use data to decide whether a claim about a population is likely true or not.<br>
**The basic idea** <br>
You start with two competing statements:<br>
**Null hypothesis (H‚ÇÄ):** the ‚Äústatus quo‚Äù or default assumption<br>
Nothing interesting is happening<br>
Example: ‚ÄúThe average exam score is 70.‚Äù<br>
**Alternative hypothesis (H‚ÇÅ or H‚Çê):** what you‚Äôre testing for<br>
Something has changed or is different<br>
Example: ‚ÄúThe average exam score is higher than 70.‚Äù

#### 2.What is the null hypothesis, and how does it differ from the alternative hypothesis?

**Answer:-** The null hypothesis and the alternative hypothesis are the two opposing claims that frame a hypothesis test. They play very different roles.<br>
**Null hypothesis (H‚ÇÄ):** <br>
The null hypothesis is the default assumption. It says that there is no effect, no difference, or no change in the population.<br>
‚Ä¢It‚Äôs what we assume is true unless the data strongly suggests otherwise.<br>
‚Ä¢It usually includes an equality ( =, ‚â§, or ‚â• ).<br>
**Alternative hypothesis (H‚ÇÅ or H‚Çê):** <br>
The alternative hypothesis is what you‚Äôre actually looking for evidence in favor of. It says that there is an effect, a difference, or a change.<br>
‚Ä¢It contradicts the null hypothesis.<br>
‚Ä¢It uses inequality ( ‚â†, <, or > ).

#### 3.What is the significance level in hypothesis testing, and why is it important?

**Answer:-** The significance level (written as Œ±, alpha) is the threshold for how strong the evidence must be before you‚Äôre willing to reject the null hypothesis.<br>
**Why the significance level is important** <br>
**1. It controls false positives** <br>
Without Œ±, you‚Äôd have no rule for deciding when evidence is ‚Äústrong enough.‚Äù<br>
**2. It forces you to decide before seeing the data** <br>
Choosing Œ± ahead of time prevents bias and cherry-picking results.<br>
**3. It reflects how serious mistakes are** <br>
Different fields tolerate different risks:<br>
‚Ä¢Medicine: Œ± = 0.01 or lower (mistakes are costly)<br>
‚Ä¢Social sciences: Œ± = 0.05 (common balance)<br>
‚Ä¢Exploratory research: sometimes Œ± = 0.10<br>
**4. It gives results a clear decision rule**
Everyone can interpret results the same way, not just based on intuition.

#### 4.What does a p-value represent in hypothesis testing?

**Answer:-** The p-value is the probability of observing results at least as extreme as the ones you got, assuming the null hypothesis (H‚ÇÄ) is true.

#### 5.How do you interpret the P-value in hypothesis testing?

**Answer:-** Interpreting a p-value is about judging how much evidence your data provides against the null hypothesis (H‚ÇÄ)‚Äînot about proving anything true or false.<br>
**The right way to interpret a p-value** <br>
A p-value answers this question:<br>
*If the null hypothesis were true, how likely is it to observe results this extreme (or more)?* <br>
From there:<br>
‚Ä¢Small p-value ‚Üí data is unlikely under H‚ÇÄ ‚Üí strong evidence against H‚ÇÄ<br>
‚Ä¢Large p-value ‚Üí data is plausible under H‚ÇÄ ‚Üí weak evidence against H‚ÇÄ

#### 6.What are Type1 and Type2 errors in hypothesis testing?

**Answer:-** Type I and Type II errors are the two ways hypothesis testing can go wrong. They describe which kind of mistake you make when deciding about the null hypothesis.<br>
**Type I error (false positive)** <br>
A Type I error happens when you reject the null hypothesis even though it‚Äôs actually true.<br>
‚Ä¢You think there is an effect, but there isn‚Äôt.<br>
‚Ä¢Probability of a Type I error = Œ± (the significance level).<br>
**Type II error (false negative)** <br>
A Type II error happens when you fail to reject the null hypothesis even though it‚Äôs false.<br>
‚Ä¢You miss a real effect.<br>
‚Ä¢Probability of a Type II error = Œ≤.

#### 7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

**Answer:-** The difference between a one-tailed and a two-tailed test comes down to what kinds of departures from the null hypothesis you‚Äôre looking for‚Äîand where you place the ‚Äúrejection region.‚Äù<br>
**One-tailed test:** <br>
A one-tailed test looks for an effect in one specific direction only.<br>
‚Ä¢The alternative hypothesis is directional (greater than or less than).<br>
‚Ä¢All of the significance level (Œ±) is placed in one tail of the distribution.<br>
**Two-tailed test:** <br>
A two-tailed test looks for an effect in either direction.<br>
‚Ä¢The alternative hypothesis is non-directional (not equal to).<br>
‚Ä¢The significance level (Œ±) is split between both tails of the distribution.

#### 20.What is an ANOVA test and what are its assumptions?

**Answer:-** ANOVA (Analysis of Variance) is a hypothesis test used to determine whether three or more population means are statistically different from each other. Instead of comparing means directly, ANOVA compares variability‚Äîwhich turns out to be a cleaner, safer way to test multiple groups at once.<br>
**Assumptions of ANOVA** <br>
1. Independence:- <br>
‚Ä¢Observations are independent within and across groups<br>
‚Ä¢Usually ensured by proper study design (random sampling/assignment)<br>
2. Normality:- <br>
‚Ä¢The dependent variable is normally distributed within each group<br>
‚Ä¢ANOVA is fairly robust to moderate violations, especially with larger samples<br>
3. Homogeneity of variances:- <br>
‚Ä¢All groups have approximately equal population variances<br>
‚Ä¢Often checked using Levene‚Äôs test or Bartlett‚Äôs test

#### 3.Write a python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pmf(x, p):
    """
    Calculate the PMF of a Bernoulli distribution.

    Parameters:
    x : int (0 or 1)
    p : float (probability of success, 0 <= p <= 1)

    Returns:
    float : P(X = x)
    """
    if x not in [0, 1]:
        return 0
    return (p ** x) * ((1 - p) ** (1 - x))
```

```python
p = 0.7  # probability of success

print("P(X = 1):", bernoulli_pmf(1, p))
print("P(X = 0):", bernoulli_pmf(0, p))
```

#### 9.Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_z_scores(data):
    """
    Calculate Z-scores for a dataset and plot them.

    Parameters:
    data : array-like
        Input dataset (list, numpy array, etc.)

    Returns:
    z_scores : numpy array
        Z-scores of the dataset
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data, ddof=0)  

    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(8, 4))
    plt.plot(z_scores, 'o', color='blue', label='Z-scores')
    plt.axhline(0, color='red', linestyle='--', label='Mean (0)')
    plt.axhline(1, color='green', linestyle='--', label='+1 Std Dev')
    plt.axhline(-1, color='green', linestyle='--', label='-1 Std Dev')
    plt.xlabel("Observation Index")
    plt.ylabel("Z-score")
    plt.title("Z-scores of the Dataset")
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_scores
```

```python
# Simulate a dataset
data = np.random.normal(50, 10, 20)  # mean=50, std=10, 20 samples

# Calculate and plot Z-scores
z_scores = plot_z_scores(data)

# Print Z-scores
print("Z-scores:", z_scores)
```

#### 10.Implement the Central Limit Theorem(CLT) using python for a non-normal distribution.

#### 12.Write a python function to calculate and plot the standard normal distribution (mean=0,std=1).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal():
    """
    Plots the Standard Normal Distribution (mean=0, std=1)
    """
    
    # Define range for x-axis
    x = np.linspace(-4, 4, 1000)
    
    # Calculate standard normal PDF
    y = norm.pdf(x, 0, 1)
    
    # Plot the distribution
    plt.figure(figsize=(8,5))
    plt.plot(x, y, label='Standard Normal Distribution', color='blue')
    
    # Fill area under curve
    plt.fill_between(x, y, alpha=0.2, color='blue')
    
    # Formatting
    plt.title("Standard Normal Distribution (Œº=0, œÉ=1)")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.axvline(0, color='black', linestyle='--', label='Mean (0)')
    plt.legend()
    plt.grid(alpha=0.3)
    
    plt.show()

# Call the function
plot_standard_normal()
```

#### 13.Generate a random variables and calculate their corresponding probabilities using the binomial distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def binomial_simulation(n=10, p=0.5, num_samples=20):
    """
    n  : number of trials
    p  : probability of success
    num_samples : number of random variables to generate
    """
    
    # Step 1: Generate random variables from Binomial distribution
    random_vars = np.random.binomial(n, p, num_samples)
    
    # Step 2: Calculate probabilities using PMF
    probabilities = binom.pmf(random_vars, n, p)
    
    # Display results
    print("Generated Random Variables (X):")
    print(random_vars)
    print("\nCorresponding Probabilities P(X = x):")
    print(probabilities)
    
    # Step 3: Plot full Binomial distribution
    x = np.arange(0, n+1)
    pmf = binom.pmf(x, n, p)
    
    plt.figure(figsize=(8,5))
    plt.stem(x, pmf, basefmt=" ")
    plt.title(f"Binomial Distribution (n={n}, p={p})")
    plt.xlabel("Number of Successes (x)")
    plt.ylabel("Probability P(X=x)")
    plt.grid(alpha=0.3)
    plt.show()

# Example usage
binomial_simulation(n=10, p=0.5, num_samples=20)
```

#### 14.Write a python program to calculate the Z-score for a given data point and compare it to standard normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    """
    x    : data point
    mean : population mean
    std  : population standard deviation
    """
    
    # Step 1: Calculate Z-score
    z = (x - mean) / std
    
    # Step 2: Calculate probability from standard normal distribution
    p_value = norm.cdf(z)   # P(Z <= z)
    
    print(f"Data Point (x): {x}")
    print(f"Mean (Œº): {mean}")
    print(f"Standard Deviation (œÉ): {std}")
    print(f"Z-score: {z:.4f}")
    print(f"P(Z ‚â§ {z:.4f}) = {p_value:.4f}")
    
    # Step 3: Plot standard normal distribution
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals)
    
    plt.figure(figsize=(8,5))
    plt.plot(x_vals, y_vals, label="Standard Normal Distribution")
    plt.fill_between(x_vals, y_vals, where=(x_vals <= z), alpha=0.3)
    
    plt.axvline(z, color='red', linestyle='--', label=f"Z = {z:.2f}")
    plt.title("Z-score on Standard Normal Distribution")
    plt.xlabel("Z")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()

# Example usage
calculate_z_score(x=85, mean=70, std=10)
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

def z_test(sample, mu_0, sigma, alpha=0.05):
    """
    sample : sample dataset (array-like)
    mu_0   : hypothesized population mean
    sigma  : known population standard deviation
    alpha  : significance level
    """
    
    n = len(sample)
    sample_mean = np.mean(sample)
    
    # Step 1: Calculate Z-statistic
    z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))
    
    # Step 2: Calculate p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    
    # Step 3: Critical value
    z_critical = norm.ppf(1 - alpha/2)
    
    # Step 4: Decision
    reject = abs(z_stat) > z_critical
    
    print("Sample Mean:", round(sample_mean, 4))
    print("Z-statistic:", round(z_stat, 4))
    print("Critical Z-value:", round(z_critical, 4))
    print("P-value:", round(p_value, 6))
    
    if reject:
        print("Decision: Reject the Null Hypothesis (H0)")
    else:
        print("Decision: Fail to Reject the Null Hypothesis (H0)")
    
    # Plot
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)
    
    plt.figure(figsize=(8,5))
    plt.plot(x, y, label="Standard Normal Distribution")
    
    # Shade rejection regions
    plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.3)
    plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.3)
    
    plt.axvline(z_stat, color='black', linestyle='--', label=f"Z = {z_stat:.2f}")
    plt.title("Z-Test Hypothesis Testing")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()

np.random.seed(42)
# Sample of 50 observations
sample_data = np.random.normal(loc=105, scale=15, size=50)
# Hypothesis test
z_test(sample=sample_data, mu_0=100, sigma=15, alpha=0.05)
```

#### 16.Create a confidence interval for a dataset usinf python and interpret the result.

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    data       : sample dataset (array-like)
    confidence : confidence level (default 95%)
    """
    
    n = len(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)  
    se = std / np.sqrt(n)       
    
    # t critical value
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)
    
    margin_of_error = t_critical * se
    
    lower = mean - margin_of_error
    upper = mean + margin_of_error
    
    print(f"Sample Size (n): {n}")
    print(f"Sample Mean: {mean:.4f}")
    print(f"Sample Std Dev: {std:.4f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.4f}, {upper:.4f})")
    
    return lower, upper

np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)

confidence_interval(sample_data, confidence=0.95)
```

#### 3.Implement a one-sample Z-test using python to compare the sample mean with the population mean.

```python
import numpy as np
import math
from scipy.stats import norm

def one_sample_z_test(sample_data, population_mean, population_std, alpha=0.05, alternative='two-sided'):
    """
    Perform a one-sample Z-test.
    
    Parameters:
        sample_data (array-like): Sample observations
        population_mean (float): Known population mean (Œº)
        population_std (float): Known population standard deviation (œÉ)
        alpha (float): Significance level
        alternative (str): 'two-sided', 'greater', or 'less'
    """
    
    # Convert to numpy array
    sample_data = np.array(sample_data)
    
    # Sample statistics
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    
    # Calculate Z statistic
    z_stat = (sample_mean - population_mean) / (population_std / math.sqrt(n))
    
    # Calculate p-value
    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_stat)
    elif alternative == 'less':
        p_value = norm.cdf(z_stat)
    else:
        raise ValueError("alternative must be 'two-sided', 'greater', or 'less'")
    
    # Print results
    print("Sample Mean:", round(sample_mean, 4))
    print("Z-Statistic:", round(z_stat, 4))
    print("P-Value:", round(p_value, 4))
    
    # Interpretation
    if p_value < alpha:
        print("\nReject the null hypothesis.")
    else:
        print("\nFail to reject the null hypothesis.")
        
    return z_stat, p_value

np.random.seed(0)

# Simulated sample data
sample = np.random.normal(loc=105, scale=15, size=36)

# Known population parameters
population_mean = 100
population_std = 15

one_sample_z_test(sample, population_mean, population_std)
```

#### 5.Create a python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math

def visualize_type1_type2(mu0, mu1, sigma, n, alpha=0.05):
    """
    Visualizes Type I and Type II errors for a two-tailed Z-test.
    
    Parameters:
        mu0 : Null hypothesis mean
        mu1 : True mean under alternative hypothesis
        sigma : Population standard deviation
        n : Sample size
        alpha : Significance level
    """
    
    # Standard error
    se = sigma / math.sqrt(n)
    
    # Critical Z values
    z_critical = norm.ppf(1 - alpha/2)
    
    # Convert Z critical values to sample mean scale
    critical_lower = mu0 - z_critical * se
    critical_upper = mu0 + z_critical * se
    
    # X values (sample mean scale)
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    
    # Null and Alternative distributions
    null_dist = norm.pdf(x, mu0, se)
    alt_dist = norm.pdf(x, mu1, se)
    
    # Compute Beta (Type II error)
    beta = norm.cdf(critical_upper, mu1, se) - norm.cdf(critical_lower, mu1, se)
    power = 1 - beta
    
    # Plot distributions
    plt.figure(figsize=(10,6))
    plt.plot(x, null_dist, label="Null Distribution (H0)", color="blue")
    plt.plot(x, alt_dist, label="Alternative Distribution (H1)", color="green")
    
    # Shade Type I error (alpha)
    plt.fill_between(x, null_dist, where=(x <= critical_lower), color='red', alpha=0.5)
    plt.fill_between(x, null_dist, where=(x >= critical_upper), color='red', alpha=0.5, label="Type I Error (Œ±)")
    
    # Shade Type II error (beta)
    plt.fill_between(x, alt_dist, where=(x >= critical_lower) & (x <= critical_upper),
                     color='orange', alpha=0.5, label="Type II Error (Œ≤)")
    
    # Critical lines
    plt.axvline(critical_lower, color='black', linestyle='--')
    plt.axvline(critical_upper, color='black', linestyle='--')
    
    plt.title("Type I and Type II Errors (Two-Tailed Z-Test)")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()
    
    print(f"Significance Level (Œ±): {alpha}")
    print(f"Type II Error (Œ≤): {beta:.4f}")
    print(f"Power (1 - Œ≤): {power:.4f}")

# Example usage
visualize_type1_type2(mu0=100, mu1=105, sigma=15, n=36, alpha=0.05)
```

#### 6.Write a python program to perform an independent T-test and interpret the results.

```python
import numpy as np
from scipy.stats import ttest_ind

def independent_t_test(group1, group2, alpha=0.05):
    """
    Perform an independent two-sample t-test.
    
    Parameters:
        group1 (array-like): First sample
        group2 (array-like): Second sample
        alpha (float): Significance level
    """
    
    group1 = np.array(group1)
    group2 = np.array(group2)
    
    t_stat, p_value = ttest_ind(group1, group2, equal_var=False)
    
    print("Group 1 Mean:", round(np.mean(group1), 4))
    print("Group 2 Mean:", round(np.mean(group2), 4))
    print("T-Statistic:", round(t_stat, 4))
    print("P-Value:", round(p_value, 4))
    
    if p_value < alpha:
        print("\nReject the null hypothesis.")
        print("There is a statistically significant difference between the two group means.")
    else:
        print("\nFail to reject the null hypothesis.")
        print("There is no statistically significant difference between the two group means.")
    
    return t_stat, p_value

np.random.seed(42)

group1 = np.random.normal(loc=75, scale=10, size=30)
group2 = np.random.normal(loc=80, scale=12, size=30)

independent_t_test(group1, group2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

def paired_t_test_visualization(before, after, alpha=0.05):
    """
    Performs a paired t-test and visualizes the comparison.
    
    Parameters:
        before : array-like (measurements before treatment)
        after  : array-like (measurements after treatment)
        alpha  : significance level
    """
    
    before = np.array(before)
    after = np.array(after)
    
    t_stat, p_value = ttest_rel(before, after)
    
    differences = after - before
    
    print("Mean Before:", round(np.mean(before), 4))
    print("Mean After :", round(np.mean(after), 4))
    print("Mean Difference:", round(np.mean(differences), 4))
    print("T-Statistic:", round(t_stat, 4))
    print("P-Value:", round(p_value, 4))
    
    if p_value < alpha:
        print("\nReject the null hypothesis.")
        print("There is a statistically significant difference between before and after measurements.")
    else:
        print("\nFail to reject the null hypothesis.")
        print("There is no statistically significant difference between before and after measurements.")
    
    plt.figure(figsize=(10,6))
    
    for i in range(len(before)):
        plt.plot([1, 2], [before[i], after[i]], marker='o', color='gray', alpha=0.6)
    
    plt.scatter(1, np.mean(before), color='blue', s=100, label='Mean Before')
    plt.scatter(2, np.mean(after), color='red', s=100, label='Mean After')
    
    plt.xticks([1, 2], ['Before', 'After'])
    plt.title('Paired Sample T-Test Comparison')
    plt.ylabel('Measurement Value')
    plt.legend()
    plt.grid(True)
    plt.show()

np.random.seed(1)

before = np.random.normal(loc=70, scale=8, size=20)
after = before + np.random.normal(loc=5, scale=5, size=20)  

paired_t_test_visualization(before, after)
```

```python
import numpy as np
import math
from scipy.stats import norm, t, ttest_1samp

def compare_z_test_t_test():
    
    np.random.seed(42)
    
    population_mean = 100
    population_std = 15   
    sample_size = 25      
    alpha = 0.05
    sample = np.random.normal(loc=105, scale=population_std, size=sample_size)
    
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    
    print("Sample Mean:", round(sample_mean, 4))
    print("Sample Std Dev:", round(sample_std, 4))
    print()
    
    z_stat = (sample_mean - population_mean) / (population_std / math.sqrt(sample_size))
    z_p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    
    print("----- Z-Test (œÉ known) -----")
    print("Z-Statistic:", round(z_stat, 4))
    print("P-Value:", round(z_p_value, 4))
    
    if z_p_value < alpha:
        print("Decision: Reject H0")
    else:
        print("Decision: Fail to reject H0")
    
    print()
    
    t_stat, t_p_value = ttest_1samp(sample, population_mean)
    
    print("----- T-Test (œÉ unknown) -----")
    print("T-Statistic:", round(t_stat, 4))
    print("P-Value:", round(t_p_value, 4))
    
    if t_p_value < alpha:
        print("Decision: Reject H0")
    else:
        print("Decision: Fail to reject H0")
    
    print("\nComparison Summary:")
    print("Z-test uses known population standard deviation.")
    print("T-test uses sample standard deviation and t-distribution.")
    print("For small samples, T-test is more appropriate.")

compare_z_test_t_test()
```

#### 9.Write a python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy import stats

def confidence_interval_mean(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
        data (list or array-like): Sample data.
        confidence (float): Confidence level (default is 0.95 for 95%).

    Returns:
        (mean, lower_bound, upper_bound)
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    
    # t critical value for two-tailed test
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)
    
    margin_of_error = t_critical * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    
    return mean, lower_bound, upper_bound

sample = [12, 15, 14, 10, 13, 16, 12, 14]
mean, lower, upper = confidence_interval_mean(sample)

print(f"Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

#### 10.Write a python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
from scipy import stats

def margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a sample mean.

    Parameters:
        data (list or array-like): Sample data
        confidence (float): Confidence level (default 0.95)

    Returns:
        float: Margin of error
    """
    data = np.array(data)
    n = len(data)
    
    sample_std = np.std(data, ddof=1)
    
    std_error = sample_std / np.sqrt(n)
    
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)
    
    moe = t_critical * std_error
    
    return moe

if __name__ == "__main__":
    sample_data = [10, 12, 9, 11, 13, 12, 8, 10]
    confidence_level = 0.95
    
    moe = margin_of_error(sample_data, confidence_level)
    print(f"Margin of Error (95% CI): {moe:.4f}")
```

#### 11.Implement a Bayesian inference method usinf Bayes'theorem in python and explain the process.

```python
def bayes_theorem(prior, likelihood, false_positive_rate):
    """
    Computes posterior probability using Bayes' Theorem.

    prior: P(H)
    likelihood: P(E|H)
    false_positive_rate: P(E|not H)
    """

    # Probability of not having disease
    prior_not = 1 - prior

    # Total probability of evidence
    evidence = (likelihood * prior) + (false_positive_rate * prior_not)

    # Posterior probability
    posterior = (likelihood * prior) / evidence

    return posterior


# Given probabilities
prior = 0.01                 # P(Disease)
likelihood = 0.99            # P(Positive | Disease)
false_positive_rate = 0.05   # P(Positive | No Disease)

posterior = bayes_theorem(prior, likelihood, false_positive_rate)

print(f"Probability of having disease given positive test: {posterior:.4f}")
```

#### 13.Write a python program to calculate the expected frequencies for a Chi-square test based on observed data.

```python
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test.

    Parameters:
    observed (2D list or numpy array): Contingency table of observed counts

    Returns:
    expected (numpy array): Expected frequency table
    """
    
    observed = np.array(observed)
   
    row_totals = observed.sum(axis=1)
    
    col_totals = observed.sum(axis=0)
    
    grand_total = observed.sum()
    
    expected = np.outer(row_totals, col_totals) / grand_total
    
    return expected

observed_data = [
    [30, 20],
    [20, 30]
]

expected = calculate_expected_frequencies(observed_data)

print("Observed Frequencies:\n", np.array(observed_data))
print("\nExpected Frequencies:\n", expected)
```

#### 19.Write a python function to check the assumptions (normality,independence, and equal variance) for ANOVA .

```python
import numpy as np
import pandas as pd
from scipy.stats import shapiro, levene, bartlett
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(df, value_col, group_col, alpha=0.05):
    """
    Checks the main assumptions for One-Way ANOVA:
    1. Normality within each group (Shapiro-Wilk test)
    2. Homogeneity of variance (Levene and Bartlett tests)
    3. Independence (visual check with scatterplot)

    Parameters:
    - df: pandas DataFrame containing the data
    - value_col: column name of numeric values
    - group_col: column name of groups
    - alpha: significance level for statistical tests

    Returns:
    - Dictionary with test results
    """

    groups = df[group_col].unique()
    results = {}

    print("\n--- 1Ô∏è‚É£ Normality Test (Shapiro-Wilk) ---")
    normality_pass = True
    for g in groups:
        data = df[df[group_col]==g][value_col]
        stat, p = shapiro(data)
        results[f"Normality_{g}"] = {'W-stat': stat, 'p-value': p, 'Normal?': p > alpha}
        print(f"Group {g}: W={stat:.3f}, p={p:.3f} ‚Üí {'Normal' if p>alpha else 'Not normal'}")
        if p <= alpha:
            normality_pass = False

    print("\n--- 2Ô∏è‚É£ Homogeneity of Variance ---")
    
    samples = [df[df[group_col]==g][value_col] for g in groups]
    stat_levene, p_levene = levene(*samples)
    results['Levene'] = {'Statistic': stat_levene, 'p-value': p_levene, 'Equal Variance?': p_levene > alpha}
    print(f"Levene Test: W={stat_levene:.3f}, p={p_levene:.3f} ‚Üí {'Equal variance' if p_levene>alpha else 'Unequal variance'}")

    stat_bart, p_bart = bartlett(*samples)
    results['Bartlett'] = {'Statistic': stat_bart, 'p-value': p_bart, 'Equal Variance?': p_bart > alpha}
    print(f"Bartlett Test: œá¬≤={stat_bart:.3f}, p={p_bart:.3f} ‚Üí {'Equal variance' if p_bart>alpha else 'Unequal variance'}")

    print("\n--- 3Ô∏è‚É£ Independence Check (Visual) ---")
    plt.figure(figsize=(8,6))
    sns.scatterplot(x=np.arange(len(df)), y=value_col, hue=group_col, data=df, palette='Set2')
    plt.title("Visual Check of Independence (Random order of samples recommended)")
    plt.xlabel("Observation Index")
    plt.ylabel(value_col)
    plt.show()
    
    print("Note: Independence is best ensured by study design; visual check only suggests potential patterns.")

    results['Normality_pass'] = normality_pass
    results['Variance_pass'] = p_levene > alpha

    return results
```

#### 20.Perform a two-way ANOVA test using python to study the interation between two factors and visualize the results.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols

np.random.seed(42)

fertilizer = ['A', 'B']
watering = ['Low', 'High']

data = []

for f in fertilizer:
    for w in watering:
    
        if f=='A' and w=='Low':
            growth = np.random.normal(20, 2, 10)
        elif f=='A' and w=='High':
            growth = np.random.normal(25, 2, 10)
        elif f=='B' and w=='Low':
            growth = np.random.normal(22, 2, 10)
        else:  # B & High
            growth = np.random.normal(30, 2, 10)
        
        for g in growth:
            data.append([f, w, g])

df = pd.DataFrame(data, columns=['Fertilizer', 'Watering', 'Growth'])

model = ols('Growth ~ C(Fertilizer) + C(Watering) + C(Fertilizer):C(Watering)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("Two-Way ANOVA Table:\n")
print(anova_table)

alpha = 0.05
for factor in ['C(Fertilizer)', 'C(Watering)', 'C(Fertilizer):C(Watering)']:
    p = anova_table.loc[factor, 'PR(>F)']
    if p < alpha:
        print(f"{factor} ‚Üí Significant effect (p={p:.3f})")
    else:
        print(f"{factor} ‚Üí No significant effect (p={p:.3f})")

plt.figure(figsize=(8,6))
sns.pointplot(x='Fertilizer', y='Growth', hue='Watering', data=df,
              markers=['o', 's'], capsize=.1, errwidth=1, palette='Set2')
plt.title('Two-Way ANOVA Interaction Plot')
plt.ylabel('Plant Growth')
plt.show()
```
